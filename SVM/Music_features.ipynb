### les Kernels

###### Paramètre
    
    - C : Contrôle le compromis entre la classification correcte des points et limite de décision.
       . C, petit: Le coût (pénalité) d'une mauvaise classification par le modèle sera faible
       . C, grand: Rend le coût d'une mauvaise classification élevé, forçe le modèle à classer les données d'entrées beaucoup plus finement .
    
    - gamma : Contrôle l’influence d’un seul ensemble d’entraînement sur toute la suite.
       . gamma, grand : les points de données plus proches ont un poids plus élevés
       . gamma, petit : grande portée (solution plus généralisée)
